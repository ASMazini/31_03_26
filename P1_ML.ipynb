```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression


dados = {
    "Tinta Acrílica": [120, 150, 130, 180, 200, 210, 190, 220, 210, 250, 300, 400],
    "Tinta Esmalte": [80, 100, 90, 120, 130, 140, 150, 160, 170, 180, 190, 200],
    "Tinta Látex": [200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300, 310],
    "Tinta Spray": [60, 70, 65, 80, 85, 90, 95, 100, 105, 110, 120, 130],
    "Tinta PVA": [150, 160]
}

plt.figure(figsize=(15, 10))

for i, (produto, vendas) in enumerate(dados.items(), 1):

    meses = np.array(range(1, len(vendas) + 1)).reshape(-1, 1)
    vendas_reais = np.array(vendas)

    modelo = LinearRegression()
    modelo.fit(meses, vendas_reais)

    meses_futuros = np.array(range(1, 16)).reshape(-1, 1)
    vendas_previstas = modelo.predict(meses_futuros)

    plt.subplot(3, 2, i)

    plt.scatter(meses, vendas_reais, color='blue', label='Vendas Reais', s=50)

    plt.plot(meses_futuros, vendas_previstas, color='red', linestyle='--', label='Linha de Tendência (Previsão)')

    plt.title(f"{produto}")
    plt.xlabel("Meses")
    plt.ylabel("Vendas (Unidades)")
    plt.xticks(range(1, 16))
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()


plt.tight_layout()
plt.show()
```
